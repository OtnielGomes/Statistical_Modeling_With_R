# <font color = 'lightgreen'>Creating a Statistical Model</font>
***

#### Rules for creating statistical models

* Y(Response) ~ X(Explanatory)

* Y(Response) ~ X1(Explanatory) + X2(Explanatory) + X3(Explanatory) ...

* **p-value indicates the significance of the model or variable**

* If p-value < 0.05, then your model is correct and the variable is significant.

* If p-value > 0.05 there is no expected effect

***
# <font color = 'lightblue'> Statistical model with $ vala $ variable</font>
***

```r
Model_vala <- lm(airq ~ vala, data = Airq)
Model_vala
```

```r
summary(Model_vala)
```

#### Considering that the value of p_value: 0.07938 is greater than the significance: 0.05:
* ### The vala variable did not influence air quality in cities

```r
plot(airq ~ vala, 
     data = Airq, 
     xlab = 'Value of companies in the city', 
     ylab = 'Air quality',
     main = 'Air quality analysis',
     pch = 1,
     col = 'blue',
     cex.lab = 1.3,
     cex = 2
    )
curve(96.451419 + 0.001969 * x, 
      add = TRUE, col = 'darkblue', 
      lwd = 2, lty = 1
     )
```

#### Points to check:
* Are the models the same?
* If p > 0.05 There is no difference between the models, and then we will continue with the simplest model
* If p < 0.05, then the models are different and the variable should not be removed from the model
