## <font color = 'lightgreen'>Creating a Statistical Model</font>
***

Y(Resposta) ~ X(Explicativa)

Y(Resposta) ~ X1(Explicativa) + X2(Explicativa) + X3(Explicativa) ...

```r
model_1 <- lm(airq ~ vala, data = Airq)
model_1
```

```r
summary(model_1)
```

**p-valor indica a significancia do modelo ou da variavél**

Se p-valor < 0.05 então compreende que seu modelo está acertivo, a varivél é significativa.

Se p-valor > 0.05 não existe o efeito esperado

A varivél vala não influenciou a qualidade do ar nas cidades

## <font color = 'lightgreen'>Multiple Regression</font>
***

Variavél continua e uma série de variaveis continuas ou não que podem explicar ou não a variavél resposta

```r
model_multiple <- lm(airq ~ vala + coas, data = Airq)
model_multiple
```

```r
summary(model_multiple)
```

```r
plot(airq ~ vala, data = Airq, 
     xlab = 'Valor das Empresas', 
     ylab = 'Qualidade do Ar',
     main = 'Analise da qualidade do ar',
     pch = 1, col = 'blue', 
     cex.lab = 1.3, cex = 2, 
     ylim = c(50, 170))
curve(1.171e+02 + 1.999e-03 * x, add = TRUE)# Cidade não costeira
curve(1.171e+02 + 1.999e-03 * x + -2.968e+01,lty = 2, add = TRUE)# Cidade costeira
legend('bottomright', c('Não-Costeiras', 'Costeiras'), pch = 1, lty = c(1, 2), bty = 'n')
```

### Conclusão: Então existe um efeito da posição costeira e o valor das empresas na qualidade do ar

#### A qualidade do ar é afetada tanto pelo valor das empresas e pela posição costeira das cidades. Quando maior o valor das empresas, pior a qualidade do ar das cidades. Além disso, as cidades não costeiras apresentam uma qualidade do ar pior que as cidades costeiras.

```r
model_multiple_2 <- lm(airq ~ vala + coas + dens, data = Airq)
model_multiple_2
```

```r
summary(model_multiple_2)
```

## <font color = 'lightgreen'>Contrast of Models</font>
***

### Comparar um modelo completo com um modelo sem a variável em questão 

```r
model_complete <- lm(airq ~ vala + coas + dens, data = Airq)
model_complete
```

```r
model_test <- lm(airq ~ vala + coas, data = Airq)
model_test
```

#### Pontos a serem verificados:
* Os modelos são iguais?
* Se p > 0.05 Não existe diferença entre os modelos,e então continuaremos com o modelo mais simples
* Se p < 0.05 Então os m,odelos são diferentes e a variavél não deve ser retirada do modelo

```r
anova(model_complete, model_test)
```

```r
plot(airq ~ vala, data = Airq, 
     xlab = 'Valor das Empresas', 
     ylab = 'Qualidade do Ar',
     main = 'Analise da qualidade do ar',
     pch = 1, 
     col = 'blue', 
     cex.lab = 1.3, 
     cex = 2, 
     ylim = c(50, 170))
curve(1.171e+02 + 1.999e-03 * x, add = TRUE, col = 'darkblue', lwd = 1.4)# Cidade não costeira
curve(1.171e+02 + 1.999e-03 * x + -2.968e+01,lty = 2, add = TRUE, col = 'darkblue', lwd = 1.4)# Cidade costeira
legend('bottomright', c('Não-Costeiras', 'Costeiras'), pch = 1, lty = c(1, 2), bty = 'n', col = c('darkblue', 'darkblue'))
```

### Conclusão: Entende-se que não existe diferença entre os dois modelos, pois o p valor é > que 0.05, então optaremos em seguir com o model_test pois entendemos que não perderemos nenhuma informação no processo.

### O que afeta a qualidade do ar nas cidades?
* Valor das empresas:  Quanto maior o valor pior a qualidade do Ar
* Posição costeira: As cidades costeriras apresenteam uma melhor qualidade do Ar. Já as cidades não costeira não apresentam uma qualidade do Ar inferior
